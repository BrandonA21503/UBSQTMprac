```python
#import packages
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Import the dataset
who = pd.read_csv("all_data.csv")

#Find all the countries included in the dataset
country_list = who["Country"].unique()

#Rename Life Expectancy Columns
whoo = who.rename(columns={'Life expectancy at birth (years)': "life_expectancy"})

```

```python
#Creating a scatterplot for each country
country_list = who["Country"].unique()
index = 0


for country in country_list:
        plt.figure(figsize=(15,5))
        index += 1
        who_country = whoo[whoo["Country"] == country]
        plt.subplot(2,3, index)
        plt.plot(who_country["life_expectancy"], who_country["GDP"], marker = "o")
        plt.xlabel("Life Expectancy")
        plt.ylabel("GDP")
        plt.title(country + " Life Expectancy vs GDP")
        plt.show()
        
```
